```python
import pandas as pd
datos = 'https://github.com/jhoanmoreno-94/Trabajo_final/raw/refs/heads/main/results.csv'
df = pd.read_csv(datos)
print(df)
```

```python
df = df.rename(columns={"date" : "Fecha_partido",
                   "home_team" : "Nombre_equipo_local",
                   "away_team" : "Nombre_equipo_visitante",
                   "home_score" : "Resultado_equipo_local",
                   "away_score" : "Resultado_equipo_visitante",
                   "tournament" : "Nombre_torneo",
                   "city" : "Cuidad",
                   "country" : "Pais",
                   "neutral" : "Campo_neutral"
                    })
print(f"Los nombres de las columnas fueron cambiados al español, se visualizan de la siguiente manera: \n{df.columns}")
```

```python
df['Fecha_partido'] = pd.to_datetime(df['Fecha_partido'])
print(f"La columna 'Fecha partido' se actualizo su tipo de dato, ya que estaba en tipo objeto, ahora está: \n{df.dtypes}")
```

```python
print(f"El número de equipos locales son: {df['Nombre_equipo_local'].nunique()} y los nombres son: \n{df['Nombre_equipo_local'].unique()}")
```

```python
print(f"El número de equipos visitantes son: {df['Nombre_equipo_visitante'].nunique()} y los nombres son: \n{df['Nombre_equipo_visitante'].unique()}")
```
